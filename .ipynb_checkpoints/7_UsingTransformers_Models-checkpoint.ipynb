## Creating a Transformer : Using Specific model config

```python
from transformers import BertConfig, BertModel
```

### Initializing model randomly

```python
# Building the config
config = BertConfig()

## Building the model from the config and random weights
model = BertModel(config)
```

```python
print(config)
```

### Initializing model with pretrained checkpoint

```python
model = BertModel.from_pretrained('bert-base-cased')
```

The weights have been downloaded and cached (so future calls to the from_pretrained method won’t re-download them) in the cache folder, which defaults to ~/.cache/huggingface/transformers. 

```python
print(config)
```

### Saving a Model

```python
model.save_pretrained("my_model")
```

```python
!ls my_model/
```

If you take a look at the config.json file, you’ll recognize the attributes necessary to build the model architecture. 

The pytorch_model.bin file is known as the state dictionary; it contains all your model’s weights. The two files go hand in hand; the configuration is necessary to know your model’s architecture, while the model weights are your model’s parameters.
